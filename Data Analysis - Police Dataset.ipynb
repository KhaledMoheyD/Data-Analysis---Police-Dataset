### data cleaning

```python
df.head(2)
```

#### handling missing values

```python
df.count()
```

```python
df.isnull().sum()
```

#### droping columns that has more than 90% of missing values

```python
df.drop(columns=['country_name','search_type'],inplace=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.isnull().sum()
```

#### droping recording that has 9 missing values or more

```python
df.dropna(axis=0, thresh=9,inplace=True)
```

```python
df.isnull().sum()
```

#### filling the remaining missing values

```python
df['driver_gender'].fillna(df['driver_gender'].mode()[0], inplace=True)
df['driver_age'].fillna(df['driver_age'].mean(),inplace=True)
df['driver_age_raw'].fillna(df['driver_age_raw'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

### handling the stop duration as a categorical data

```python
df.stop_duration.unique()
```

```python
df.stop_duration.value_counts()
```

```python
df.drop(df[df['stop_duration']=="2"].index , inplace=True)
```

```python
df['min_stop_duration'] = df['stop_duration'].map({'0-15 Min': 0 ,'16-30 Min':16 , '30+ Min':30})
df['max_stop_duration'] = df['stop_duration'].map({'0-15 Min': 15 ,'16-30 Min':30 , '30+ Min':45})
```

```python
df['avg_stop_duration'] = (df['min_stop_duration']+df['max_stop_duration']) / 2
df
```

```python
df.drop(columns=['stop_duration'], inplace=True)
```

```python
### rearrange columns
df.insert(len(df.columns)-5,'min_stop_duration',df.pop('min_stop_duration'))
df.insert(len(df.columns)-4,'max_stop_duration',df.pop('max_stop_duration'))
df.insert(len(df.columns)-3,'avg_stop_duration',df.pop('avg_stop_duration'))
```
